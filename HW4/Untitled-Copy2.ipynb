### histogram of k that generate minimum missrate

```python
fig, ax = plt.subplots(figsize=a4_dims)
fig=sns.distplot(df.idxmin(axis = 1),bins=40)
```

```python
fig, ax = plt.subplots(figsize=a4_dims)
fig=sns.distplot(df.idxmin(axis = 0),bins=10)
```

### median of missrate w.r.t. gamma
